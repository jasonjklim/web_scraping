```python
car_4=[]
car_list = ['toyota/camry','ford/escape',  'ford/f-150', 'chevrolet/silverado-1500', 
            'honda/accord', 'ford/fusion','chevrolet/cruze', 'hyundai/sonata', 'honda/civic','honda/cr-v',
            'chevrolet/malibu', 'chevrolet/equinox', 'hyundai/elantra','ford/focus/', 'gmc/sierra-1500',
            'ford/explorer', 'toyota/rav4','jeep/grand-cherokee', 'nissan/rogue','jeep/wrangler']
for year in range(2011, 2017):
    for model in car_list:
        res = requests.get('https://www.msn.com/en-us/autos/{}/{}/'.format(model, year))
        soup = BeautifulSoup(res.content, 'lxml')

        lst=[]
        for info in soup.find_all('div', {'class':'info-container'}):
            for col in info.find('div',{'class':'value-text'}):
                lst.append(col.text.strip())
        car_dic={}
        car_dic['Name']=soup.find('div', {'class':'amy-car-content'}).find('h2').text
        car_dic['Engine']=lst[0]
        car_dic['HorsePower']=lst[1]
        car_dic['Fuel_Economy']=lst[2]
        car_dic['Body_Style'] = lst[3]
        car_dic['Transmission'] = lst[4]
        car_dic['Drive'] = lst[5]
        car_dic['Seating'] = lst[6]
        car_dic['Door'] = lst[7]
        car_dic['starting_price'] = lst[8]
        car_4.append(car_dic)

car = pd.DataFrame(car_4)
```
