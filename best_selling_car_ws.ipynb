```python
res = requests.get('http://www.goodcarbadcar.net/2017/01/usa-30-best-selling-vehicles-2016-year-end/')
```

```python
soup = BeautifulSoup(res.content, 'lxml')
```

```python
table=[]

for row in soup.find('table', {'border':"1"}).find_all('tr')[1:]:
    
    rank=[]
    for name in row.find_all('td'):
        rank.append(name.text.strip())
    table.append(rank)
        
        
car_sale_2016 = pd.DataFrame(table, columns=['Rank', 'Model', '2016', '2015', '% Change'])
```
